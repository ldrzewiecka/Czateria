```python
% Ćwiczenie 1

% Fakty dotyczące relacji rodzica
rodzic(jan, anna).
rodzic(jan, piotr).
rodzic(maria, anna).
rodzic(maria, piotr).

% Reguła dla dziadków
dziadek(X, Y) :-
    ojciec(X, Z), rodzic(Z, Y).

% Definicja płci
mezczyzna(jan).
mezczyzna(piotr).
mezczyzna(karol).
kobieta(maria).
kobieta(anna).

% Uzupełnienie relacji rodzinnych (zadanie 1.2)
matka(X, Y) :- rodzic(X, Y), kobieta(X).
ojciec(X, Y) :- rodzic(X, Y), mezczyzna(X).
corka(X, Y) :- rodzic(Y, X), kobieta(X).
syn(X, Y) :- rodzic(Y, X), mezczyzna(X).
siostra(X, Y) :- rodzic(Z, X), rodzic(Z, Y), kobieta(X), X \= Y.
brat(X, Y) :- rodzic(Z, X), rodzic(Z, Y), mezczyzna(X), X \= Y.

% Ćwiczenie 2

% Zakładamy predykaty podstawowe:
% rodzic(X, Y) – X jest rodzicem Y
% mezczyzna(X) – X jest mężczyzną
% osoba(X) – X należy do zbioru osób

% 1. kobieta(X) – X jest kobietą
kobieta(X) :- osoba(X), \+ mezczyzna(X).

% 2. ojciec(X, Y) – X jest ojcem Y
ojciec(X, Y) :- rodzic(X, Y), mezczyzna(X).

% 3. matka(X, Y) – X jest matką Y
matka(X, Y) :- rodzic(X, Y), kobieta(X).

% 4. corka(X, Y) – X jest córką Y
corka(X, Y) :- rodzic(Y, X), kobieta(X).

% 5. brat_rodzony(X, Y) – X jest rodzonym bratem Y
brat_rodzony(X, Y) :-
    rodzic(Z, X), rodzic(Z, Y),
    mezczyzna(X), X \= Y.

% 6. brat_przyrodni(X, Y) – X jest przyrodnim bratem Y
brat_przyrodni(X, Y) :-
    rodzic(Z1, X), rodzic(Z2, Y), Z1 \= Z2,
    mezczyzna(X), X \= Y.

% 7. kuzyn(X, Y) – X jest kuzynem Y
kuzyn(X, Y) :-
    rodzic(Z1, X), rodzic(Z2, Y),
    brat_rodzony(Z1, Z2).

% 8. dziadek_od_strony_ojca(X, Y) – X jest dziadkiem od strony ojca dla Y
dziadek_od_strony_ojca(X, Y) :-
    ojciec(X, Z), ojciec(Z, Y).

% 9. dziadek_od_strony_matki(X, Y) – X jest dziadkiem od strony matki dla Y
dziadek_od_strony_matki(X, Y) :-
    ojciec(X, Z), matka(Z, Y).

% 10. dziadek(X, Y) – X jest dziadkiem Y
dziadek(X, Y) :-
    dziadek_od_strony_ojca(X, Y).
dziadek(X, Y) :-
    dziadek_od_strony_matki(X, Y).

% 11. babcia(X, Y) – X jest babcią Y
babcia(X, Y) :-
    matka(X, Z), rodzic(Z, Y).

% 12. wnuczka(X, Y) – Y jest wnuczką X
wnuczka(X, Y) :-
    rodzic(X, Z), corka(Z, Y).

% 13. przodek_do2pokolenia_wstecz(X, Y) – X jest przodkiem Y do drugiego pokolenia
przodek_do2pokolenia_wstecz(X, Y) :-
    rodzic(X, Y).
przodek_do2pokolenia_wstecz(X, Y) :-
    rodzic(X, Z), rodzic(Z, Y).

% 14. przodek_do3pokolenia_wstecz(X, Y) – X jest przodkiem Y do trzeciego pokolenia
przodek_do3pokolenia_wstecz(X, Y) :-
    przodek_do2pokolenia_wstecz(X, Y).
przodek_do3pokolenia_wstecz(X, Y) :-
    rodzic(X, Z), przodek_do2pokolenia_wstecz(Z, Y).
```
